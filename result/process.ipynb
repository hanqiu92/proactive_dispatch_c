```python
import pandas as pd
import numpy as np
pd.options.mode.chained_assignment = None
```

```python
result = pd.read_csv("results.csv",names = ["strategy","congestion","demand","supply","prate","tax_cong","tax_demand",
                                            "private","revenue","cost","tax","profit","total_pax",
                                            "single_pax","shared_pax","ave_pax","ass1","ass2","ass3",
                                            "ave_dist","ave_time","public","ave_link_time","total_demand",
                                            "serviced_demand","stay_demand","serviced_rate","ave_time_offer",
                                            "ave_time_offer_s","ave_time_pickup"])

del result['public'],result['private'],result['demand']
del result['ass1'],result['ass2'],result['ass3']
```

```python
result = result.loc[result['congestion'] == 1.0,:]
del result['congestion'],result['ave_link_time']
```

```python
count = 0
compare_stra = np.zeros((3,4))
for supply in [0.25,0.5,1.0,1.5]:
    for prate in [0.5,0.6,0.7,0.8]:
        for tax in [0.0,-0.1]:
            count += 1
            temp = result.loc[(result['supply'] == supply) & (result['prate'] == prate) & (result['tax_cong'] == tax),
                              :]
            pricing_temp = np.asarray(temp.loc[(result['strategy'] == "Pricing Opt"),
                                            ['profit','ave_pax','ave_dist','ave_time']])
            full_temp = np.asarray(temp.loc[(result['strategy'] == "Full"),
                                            ['profit','ave_pax','ave_dist','ave_time']])
            single_temp = np.asarray(temp.loc[(result['strategy'] == "Single"),
                                            ['profit','ave_pax','ave_dist','ave_time']])
            shared_temp = np.asarray(temp.loc[(result['strategy'] == "Pool"),
                                            ['profit','ave_pax','ave_dist','ave_time']])
            compare_stra[0,:] += ((pricing_temp - full_temp) / pricing_temp)[0,:]
            compare_stra[1,:] += ((pricing_temp - single_temp) / pricing_temp)[0,:]
            compare_stra[2,:] += ((pricing_temp - shared_temp) / pricing_temp)[0,:]
            
compare_stra = compare_stra / count
compare_stra
```

```python
pricing_result = result.loc[result['strategy'] == "Pricing Opt",:]
del pricing_result['strategy']
pricing_result['shared_rate'] = pricing_result['shared_pax'] / pricing_result['total_pax']
pricing_result['total_shared_rate'] = pricing_result['shared_pax'] / (pricing_result['serviced_demand'] \
                                                                      + pricing_result['stay_demand'])
```

```python
pricing_result.loc[pricing_result['tax_cong'] == -0.1,['supply','prate','shared_rate','total_shared_rate']]
```

```python
pricing_result.loc[pricing_result['tax_cong'] == 0.0,['supply','prate','shared_rate','total_shared_rate']]
```

```python
pricing_result['ave_tax'] = pricing_result['tax'] / pricing_result['shared_pax'] 
pricing_result.loc[pricing_result['tax_cong'] == -0.1,['supply','prate','ave_tax']]
```

```python
pricing_result['ave_tax_veh'] = pricing_result['tax'] / pricing_result['supply'] / 200 
pricing_result.loc[pricing_result['tax_cong'] == -0.1,['supply','prate','ave_tax_veh']]
```
