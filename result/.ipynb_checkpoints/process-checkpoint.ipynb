```python
import pandas as pd
import numpy as np
pd.options.mode.chained_assignment = None
```

```python
result = pd.read_csv("results_dynamic_time_04_24.csv",
                     names = ["strategy","congestion","demand","supply","prate","tax_cong","tax_demand",
                                            "private","revenue","cost","tax","profit","total_pax",
                                            "single_pax","shared_pax","ave_pax","ave_dist","ave_time",
                                            "public","ave_link_time","total_demand",
                                            "serviced_demand","stay_demand","serviced_rate","ave_time_offer",
                                            "ave_time_offer_s","ave_time_pickup"])

del result['public'],result['private'],result['demand']
```

```python
result = result.loc[result['congestion'] == 1.0,:]
del result['congestion'],result['ave_link_time']
```

```python
count = 0
compare_stra = np.zeros((5,4))
for supply in [0.6,0.8,1.0,1.2,1.4]:
    for prate in [0.3,0.4,0.5,0.6,0.7,0.8]:
        for tax in [0.0,-0.1]:
            count += 1
            temp = result.loc[(result['supply'] == supply) & (result['prate'] == prate) & (result['tax_cong'] == tax),
                              :]
            pricing_opt_temp = np.asarray(temp.loc[(result['strategy'] == "Pricing Opt"),
                                            ['profit','ave_pax','ave_dist','ave_time']])
            full_temp = np.asarray(temp.loc[(result['strategy'] == "Full"),
                                            ['profit','ave_pax','ave_dist','ave_time']])
            single_temp = np.asarray(temp.loc[(result['strategy'] == "Single"),
                                            ['profit','ave_pax','ave_dist','ave_time']])
            shared_temp = np.asarray(temp.loc[(result['strategy'] == "Pool"),
                                            ['profit','ave_pax','ave_dist','ave_time']])
            pricing_temp = np.asarray(temp.loc[(result['strategy'] == "Pricing"),
                                            ['profit','ave_pax','ave_dist','ave_time']])
            assort_opt_temp = np.asarray(temp.loc[(result['strategy'] == "Assort Opt"),
                                            ['profit','ave_pax','ave_dist','ave_time']])
            compare_stra[0,:] += ((pricing_opt_temp - full_temp) / pricing_opt_temp)[0,:]
            compare_stra[1,:] += ((pricing_opt_temp - single_temp) / pricing_opt_temp)[0,:]
            compare_stra[2,:] += ((pricing_opt_temp - shared_temp) / pricing_opt_temp)[0,:]
            compare_stra[3,:] += ((pricing_opt_temp - assort_opt_temp) / pricing_opt_temp)[0,:]
            compare_stra[4,:] += ((pricing_opt_temp - pricing_temp) / pricing_opt_temp)[0,:]
            
compare_stra = - compare_stra / count * 100
compare_stra
```

```python
pricing_result = result.loc[result['strategy'] == "Pricing Opt",:]
del pricing_result['strategy']
pricing_result['shared_rate'] = pricing_result['shared_pax'] / pricing_result['total_pax']
pricing_result['total_shared_rate'] = pricing_result['shared_pax'] / (pricing_result['serviced_demand'] \
                                                                      + pricing_result['stay_demand'])
```

```python
pricing_result.loc[pricing_result['tax_cong'] == -0.1,['supply','prate','shared_rate','total_shared_rate']]
```

```python
pricing_result.loc[pricing_result['tax_cong'] == 0.0,['supply','prate','shared_rate','total_shared_rate']]
```

```python
pricing_result['ave_tax'] = -pricing_result['tax'] / pricing_result['shared_pax'] 
pricing_result.loc[pricing_result['tax_cong'] == -0.1,['supply','prate','ave_tax']]
```

```python
pricing_tax = pricing_result.loc[pricing_result['tax_cong'] == -0.1,:]
pricing_tax['tax_diff'] = - pricing_tax['tax'] / (pricing_tax['shared_pax'] - \
                        np.asarray(pricing_result.loc[pricing_result['tax_cong'] == 0.0,'shared_pax']))

pricing_tax.loc[:,['supply','prate','tax_diff']]

pricing_tax['tax'] / (pricing_tax['shared_pax'] - \
                        np.asarray(pricing_result.loc[pricing_result['tax_cong'] == 0.0,'shared_pax']))
```
